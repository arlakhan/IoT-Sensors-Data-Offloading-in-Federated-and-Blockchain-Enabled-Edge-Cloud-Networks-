```python
import random
import time
import hashlib
import numpy as np

# Simulate IoT Sensors (100 sensors)
NUM_SENSORS = 100
SENSOR_DATA = {f"sensor_{i}": [] for i in range(1, NUM_SENSORS + 1)}

# Blockchain to ensure integrity
class Blockchain:
    def __init__(self):
        self.chain = []

    def create_block(self, data, previous_hash):
        block = {
            'index': len(self.chain) + 1,
            'timestamp': time.time(),
            'data': data,
            'previous_hash': previous_hash,
            'hash': self.hash_block(data, previous_hash)
        }
        self.chain.append(block)
        return block

    @staticmethod
    def hash_block(data, previous_hash):
        block_string = f"{data}{previous_hash}"
        return hashlib.sha256(block_string.encode()).hexdigest()

    def get_last_block(self):
        return self.chain[-1] if self.chain else None

# Federated Learning model (Dummy model for simulation)
global_model = {"weights": np.random.rand(10)}  # 10 parameters (weights)
local_models = {}

# Simulating edge and cloud nodes
edge_nodes = [f"edge_{i}" for i in range(1, 6)]  # 5 edge nodes
cloud_nodes = [f"cloud_{i}" for i in range(1, 3)]  # 2 cloud nodes

# Federated Learning Update (simple average for demonstration)
def update_model_on_federated_learning(local_updates):
    global global_model
    updated_weights = np.mean(local_updates, axis=0)
    global_model["weights"] = updated_weights
    return global_model

# Blockchain instance
blockchain = Blockchain()
blockchain.create_block(data="Genesis Block", previous_hash="0")

# Simulate IoT sensor data generation
def generate_sensor_data():
    for sensor_id in SENSOR_DATA:
        SENSOR_DATA[sensor_id].append(random.uniform(20.0, 30.0))  # Simulate sensor data (e.g., temperature)
        time.sleep(0.1)  # Simulate periodic data

# Simulate sending data to blockchain and offloading to edge/cloud
def offload_data_to_edge_cloud():
    offload_data = {sensor_id: SENSOR_DATA[sensor_id] for sensor_id in SENSOR_DATA}
    return offload_data

# Simulate federated learning on a single node (edge or cloud)
def federated_learning_on_node(node_type, node_id, offloaded_data):
    local_data = list(offloaded_data.values())[:10]  # Simulate using data from first 10 sensors
    local_update = np.random.rand(10)  # Simulate local model update
    local_models[node_id] = local_update
    print(f"{node_type} {node_id} updated model with local data.")
    return local_update

# Aggregating federated learning updates from all nodes (edge + cloud)
def federated_learning():
    local_updates = []

    # Simulating federated learning on edge nodes
    for edge_node in edge_nodes:
        offloaded_data = offload_data_to_edge_cloud()
        local_update = federated_learning_on_node("Edge", edge_node, offloaded_data)
        local_updates.append(local_update)

    # Simulating federated learning on cloud nodes
    for cloud_node in cloud_nodes:
        offloaded_data = offload_data_to_edge_cloud()
        local_update = federated_learning_on_node("Cloud", cloud_node, offloaded_data)
        local_updates.append(local_update)

    # Update global model by aggregating local updates
    updated_model = update_model_on_federated_learning(local_updates)
    print("Global model updated:", updated_model)

# Simulate the process
def simulate_system():
    # Generate data periodically
    for _ in range(100):  # Simulate 100 rounds of data generation
        generate_sensor_data()

        # Simulate the offloading of data to edge and cloud networks
        offload_data_to_edge_cloud()

        # Simulate federated learning (edge and cloud nodes)
        federated_learning()

        time.sleep(2)  # Simulate periodic updates

if __name__ == "__main__":
    simulate_system()
```
